# Análisis de eventos deportivos con Python y Pandas

### Importar librerías

```python
import pandas as pd
import matplotlib.pyplot as pyplot
import seaborn as sns
```

### Carga de datos desde archivos .csv

```python
aficionados = pd.read_csv('workspace/aficionados.csv')
eventos = pd.read_csv('workspace/eventos.csv')
promociones = pd.read_csv('workspace/promociones.csv')
```

### Inspeccionamos los dataframes

```python
aficionados.info()
aficionados.head()
```

```python
eventos.info()
eventos.head()
```

```python
promociones.info()
promociones.head()
```

```python
df_merged = pd.merge(aficionados, eventos, on='id_evento', how='inner')
df_merged = pd.merge(df_merged, promociones, on='id_evento', how='inner')
df_merged.info()
df_merged.head()
```

```python
df_merged.tail()
```

```python
evento_mas_multitudinario = df_merged.loc[df_merged['asistentes_totales'].idxmax()]
print(evento_mas_multitudinario[['nombre_evento', 'asistentes_totales']])
```

```python
top_5_eventos = df_merged.groupby('nombre_evento')['asistentes_totales'].max().nlargest(5).reset_index()
top_5_eventos
```

```python
topes = [0, 18, 25, 35, 45, 60, 100]
rangos = ['0-18', '19-25', '26-35', '36-45', '46-60', '60+']
df_merged['rango_edad'] = pd.cut(df_merged['edad'], bins=topes, labels=rangos)
rangos = df_merged['rango_edad'].groupby(df_merged['rango_edad']).count()
rangos
```
